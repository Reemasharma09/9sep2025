```python
import os
import io
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, roc_curve, roc_auc_score, classification_report, accuracy_score
import matplotlib.pyplot as plt
import joblib
from google.colab import files

def load_csv(paths):
    for p in paths:
        if os.path.exists(p):
            return pd.read_csv(p)
    uploaded = files.upload()
    first = next(iter(uploaded.keys()))
    return pd.read_csv(io.BytesIO(uploaded[first]))

def process_dataset(df, target_col):
    zero_cols = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
    for c in zero_cols:
        if c in df.columns:
            df.loc[df[c] == 0, c] = np.nan
    for c in df.select_dtypes(include=[np.number]).columns:
        df[c].fillna(df[c].median(), inplace=True)
    for c in df.select_dtypes(include=['object','category']).columns:
        le = LabelEncoder()
        df[c] = le.fit_transform(df[c].astype(str))
    X = df.drop(columns=[target_col])
    y = df[target_col]
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    pca = PCA(n_components=2, random_state=42)
    X_pca = pca.fit_transform(X_scaled)
    plt.figure(figsize=(8,6))
    for label in np.unique(y):
        plt.scatter(X_pca[y==label,0], X_pca[y==label,1], label=str(label), alpha=0.6)
    plt.legend()
    plt.show()
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, stratify=y, test_size=0.2, random_state=42)
    models = {
        'LogisticRegression': LogisticRegression(max_iter=1000, solver='lbfgs', random_state=42),
        'SVM': SVC(kernel='rbf', C=1.5, gamma=0.3, probability=True, random_state=42),
        'DecisionTree': DecisionTreeClassifier(max_depth=6, min_samples_split=4, random_state=42),
        'RandomForest': RandomForestClassifier(n_estimators=100, max_depth=7, random_state=42),
        'GaussianNB': GaussianNB()
    }
    results = {}
    plt.figure(figsize=(8,6))
    for name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        y_proba = model.predict_proba(X_test)[:,1] if hasattr(model,'predict_proba') else np.zeros_like(y_pred)
        acc = accuracy_score(y_test, y_pred)
        auc = roc_auc_score(y_test, y_proba)
        results[name] = {'acc': acc, 'auc': auc}
        print(name, 'Accuracy:', round(acc*100,2), '%  AUC:', round(auc,3))
        print(confusion_matrix(y_test, y_pred))
        print(classification_report(y_test, y_pred, zero_division=0))
        fpr, tpr, _ = roc_curve(y_test, y_proba)
        plt.plot(fpr, tpr, label=f"{name} (AUC={auc:.3f})")
    plt.plot([0,1],[0,1],'--')
    plt.legend()
    plt.show()
    if 'RandomForest' in models:
        imp = pd.Series(models['RandomForest'].feature_importances_, index=X.columns).sort_values(ascending=False)
        imp.head(15).plot(kind='bar', figsize=(8,6))
        plt.show()
        joblib.dump(models['RandomForest'], f"{target_col}_randomforest.joblib")
        joblib.dump(scaler, f"{target_col}_scaler.joblib")
    return results

diabetes_paths = ['/mnt/data/diabetes (1).csv', '/mnt/data/diabetes.csv']
heart_paths = ['/mnt/data/heart.csv']

diabetes_df = load_csv(diabetes_paths)
heart_df = load_csv(heart_paths)

diabetes_results = process_dataset(diabetes_df, 'Outcome')
heart_results = process_dataset(heart_df, 'target')

print('Diabetes Results:', diabetes_results)
print('Heart Results:', heart_results)
```
